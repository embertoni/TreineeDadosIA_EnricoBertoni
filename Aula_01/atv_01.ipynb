- Código completo:

```python
valores_brutos: list[int] = [10, 50, 150, 20, 300]
alertas: list[int] = []

def verifica_lista(valores: list[int]):
    i: int
    for i in valores:
        if i > 100:
            alertas.append(i)

verifica_lista(valores_brutos)
if alertas == []:
    print("Sistema seguro")
else:
    print("⚠️ Revisar valores: ", alertas)
```

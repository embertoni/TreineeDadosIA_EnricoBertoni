### Otimizar Dataset:
Cenário considerado: mostrar para o cliente as relações entre os tipos de tráfego e a permanência / interesse do comprador

- Remoção das colunas irrelevantes para o contexto

```python
df = df.drop(columns=['order_id', 'payment_method', 'customer_age_days', 'delivery_days_estimated', 'late_delivery_risk', 'address_mismatch', 'high_risk_ip', 'customer_support_contacts', 'is_returned', 'is_fraud'])
df
```

```python
df.info(memory_usage='deep')
```

```python
df.memory_usage(deep=True)
```

- Conversão dos tipos das colunas

```python
df['device_type'].unique()
```

```python
df['device_type'] = df['device_type'].map({'Mobile': 0, 'Desktop': 1, 'Tablet': 2}).astype('bool')
```

```python
df['product_category'].unique()
```

```python
df['product_category'] = df['product_category'].map({'Sports': 0, 'Garden': 1, 'Electronics': 2, 'Home & Kitchen': 3, 'Pet Supplies': 4, 'Beauty': 5, 'Toys': 6, 'Fashion': 7}).astype('category')
```

```python
df['risk_label'].unique()
```

```python
df['risk_label'] = df['risk_label'].map({'Normal': 0, 'Return Risk': 1, 'Fraud Risk': 2}).astype('bool')
```

```python
df
```

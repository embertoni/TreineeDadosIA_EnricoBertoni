### Visualização Gráfica

- Importar as bibliotecas para análise e criação de gráficos comparativos (Plotly e SciPy)
    - execução de `pip install --upgrade nbformat` & `pip install statsmodels`

```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
from scipy.stats import pearsonr

pio.renderers.default = "notebook_connected"
```

Gráfico 1: Análise Comparativa de Faturamento por Canal de Tráfego
- Agrupa os dados por origem de tráfego para expor quais canais geram maior retorno financeiro e qual possui o melhor Ticket Médio

```python
traffic_insights = df.groupby('traffic_source').agg(faturamento_total=('order_value_eur', 'sum'),ticket_medio=('order_value_eur', 'mean')).reset_index()

fig_traffic = go.Figure()

fig_traffic.add_trace(go.Bar(
    x=traffic_insights['traffic_source'],
    y=traffic_insights['faturamento_total'],
    text=traffic_insights['faturamento_total'].apply(lambda x: f'€ {x:,.0f}'),
    textposition='auto',
    marker=dict(
        color=traffic_insights['ticket_medio'],
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title="Ticket Médio (€)")
    )
))

fig_traffic.update_layout(
    title='Análise Comercial: Faturamento Total e Ticket Médio por Canal de Tráfego',
    template='plotly_white',
    xaxis_title="Origem do Tráfego",
    yaxis_title="Volume Financeiro Total (EUR)",
    height=500
)

fig_traffic.show()
```

Gráfico 2: Painel Comparativo Multi-variável (Subplots) de Mercados por País
- Criar uma visualização lado a lado (Subplots) para comparar diretamente o faturamento total gerado contra a quantidade total de itens vendidos em cada país

```python
country_insights = df.groupby('country').agg(
    faturamento=('order_value_eur', 'sum'),
    quantidade_itens=('quantity', 'sum')
).sort_values(by='faturamento', ascending=False).reset_index()

fig_subplots = make_subplots(
    rows=1, cols=2, 
    subplot_titles=("Faturamento Total acumulado por País", "Quantidade Total de Itens por País"),
    shared_xaxes=False
)

fig_subplots.add_trace(
    go.Bar(
        x=country_insights['country'], 
        y=country_insights['faturamento'], 
        name='Faturamento (€)', 
        marker_color='#1f77b4',
        text=country_insights['faturamento'].round(0),
        textposition='auto'
    ),
    row=1, col=1
)

fig_subplots.add_trace(
    go.Bar(
        x=country_insights['country'], 
        y=country_insights['quantidade_itens'], 
        name='Unidades Vendidas', 
        marker_color='#2ca02c',
        text=country_insights['quantidade_itens'],
        textposition='auto'
    ),
    row=1, col=2
)

fig_subplots.update_layout(
    title_text="Painel Macro de Desempenho Comercial Internacional",
    template='plotly_white',
    showlegend=True,
    height=500
)
fig_subplots.show()
```

Gráfico 3: Validação Comercial Estatística - Descontos vs. Valor de Compra
- Mendir estatisticamente a influência das estratégias de precificação utilizando a Correlação de Pearson (`pearsonr`) para verificar se maiores taxas de desconto resultam em compras de valor maior, projetando os dados em um gráfico de dispersão com linha de tendência

```python
coef_corr, p_value = pearsonr(df['discount_rate'], df['order_value_eur'])

print(f"Coeficiente de Correlação de Pearson: {coef_corr:.4f}")
print(f"Significância Estatística (P-value): {p_value:.4e}")

fig_disp = px.scatter(
    df.sample(n=1000, random_state=42), 
    x='discount_rate', 
    y='order_value_eur',
    trendline='ols',
    title=f'Impacto Comercial de Campanhas de Desconto (Correlação: {coef_corr:.2f})',
    labels={'discount_rate': 'Taxa de Desconto Aplicada', 'order_value_eur': 'Valor Total do Pedido (EUR)'},
    opacity=0.7,
    color_discrete_sequence=['#ff7f0e']
)

fig_disp.update_layout(template='plotly_white')
fig_disp.show()
```
